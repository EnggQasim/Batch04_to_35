# Line Chart

```python
import matplotlib.pyplot as plt

x=[1,2,3]
y=[5,7,4]
plt.plot(x,y, label='First Line')

x1=[1,2,3]
x2=[10,14,12]
plt.plot(x1,x2, label="Second Line")

plt.xlabel("Plot Number")
plt.ylabel("Important var")

plt.title("Saylani Demo\n A.I class\Data Visiualization")
plt.legend()
plt.show()
```

# Bar Chart

```python
import matplotlib.pyplot as plt

x1=[2,4,6,8,10]
y1=[6,7,8,2,4]

x2=[1,3,5,7,9]
y2=[7,8,2,4,2]

plt.bar(x1,y1, label="bar1", color='r')
plt.bar(x2,y2, label="bar2", color='g')

plt.xlabel("X label")
plt.ylabel("Y label")

plt.title("Saylani Demo\n A.I class\Data Visiualization")
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt

population_ages=[22,33,44,33,22,55,66,77,88,87,98,33,22]
ids=[x
    for x,_ in enumerate(population_ages)]


plt.bar(ids,population_ages, label="Population ages")
plt.xlabel("X label")
plt.ylabel("Y label")

plt.title("Saylani Demo\n A.I class\Data Visiualization")
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt

population_ages=[22,33,44,33,22,55,66,77,88,87,98,33,22]

bins=[0,10,25,35,50,60,70,80,100,150]


plt.hist(population_ages,bins, histtype='bar', rwidth=0.8, label='Ages')
plt.xlabel("X label")
plt.ylabel("Y label")

plt.title("Saylani Demo\n A.I class\nData Visiualization")
plt.legend()
plt.show()
```

# Scater Plot

```python
import matplotlib.pyplot as plt

x=[2,4,3,4,6,5,3,5,6]
y=[1,3,5,6,6,7,7,8,9]

plt.scatter(x,y, label="data1", color='r', s=100, marker="*")

plt.title("Saylani Demo")
plt.xlabel("X")
plt.ylabel("Y")
plt.legend()

plt.show()
```

# Stackplot

```python
import matplotlib.pyplot as plt

days=[1,2,3,4,5]

reading=[3,6,8,9,9]
playing=[4,5,9,3,4]
watching=[13,7,6,8,12]
teaching=[2,15,20,4,5]

plt.plot([],[], color='c', label='days', linewidth=5)
plt.plot([],[], color='m', label='reading', linewidth=5)
plt.plot([],[], color='r', label='watching', linewidth=5)
plt.plot([],[], color='k', label='playing', linewidth=5)
plt.plot([],[], color='g', label='teaching', linewidth=5)

plt.stackplot(days, reading,watching,playing,teaching, colors=['c','m','r','k'])

plt.title("Saylani Demo")
plt.xlabel("X")
plt.ylabel("Y")
plt.legend()

plt.show()
```

# PieChart

```python
import matplotlib.pyplot as plt

slices =[7,2,2,13]
activities=["Reading","Sleeping","Playing","Teaching"]
colors=['r','c','y','k']
plt.pie(slices,
       labels=activities,
       colors=colors,
       startangle=90,
       shadow=True,
       explode=(0,0.1,0,0),
       autopct='%1.1f%%'
       )
plt.title("Saylani Demo")


plt.show()
```

# Read Data From File
file name is "abc.txt"

```python
import matplotlib.pyplot as plt
import csv
x=[]
y=[]
with open("abc.txt",  'r') as f:
    rows=csv.reader(f, delimiter=',')
    for row in rows:
        x.append(int(row[0]))
        y.append(int(row[1]))
plt.plot(x,y)
plt.show()
```

Load file with Numpy

```python
import matplotlib.pyplot as plt
import numpy as np

x,y = np.loadtxt('abc.txt', delimiter=',', unpack=True)
plt.plot(x,y)
plt.show()
```
